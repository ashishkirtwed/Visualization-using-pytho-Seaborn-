# Introduction

When exploring multi-dimensional data, a useful approach is to draw multiple instance of the same plot on different subsets on our dataset.This technique is called either "lattice" or "trellis" plotting. It is related to the idea of "small multiples". It allows a viewer to quickly extract a large amount of information about a complex dataset.

### Importing the libraries

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
```

### Conditional Small Multiples

The FacetGrid class is useful when we want to visualize the distribution of a variable or the relationship between multiple variables separately within subsets of our dataset. A FacetGrid can be drawn with up to three dimensions: row, col and hue. The first two have obvious correspondencce with the resulting array of axes; think of the hue variable as a third dimension along a depth axis.

```python
tips=sb.load_dataset("tips")
g=sb.FacetGrid(tips,col="time")
```

Initializing the grid like this sets up the matplotlib figure and axes, but doesnâ€™t draw anything on them.The main approach for visualizing data on this grid is with the FacetGrid.map() method. Provide it with a plotting function and the name(s) of variable(s) in the dataframe to plot.

```python
g.map(sb.histplot, "tip")
```

```python
g= sb.FacetGrid(tips,col="sex")
g.map(sb.scatterplot,"total_bill","tip",alpha=.7)
g.add_legend()
```

```python
g=sb.FacetGrid(tips,col="day",height=4,aspect=.5)
g.map(sb.barplot,"sex","total_bill",order=["Male","Female"])
```

Once we have drawn plot using FacetGrid.map(), we may want to adjust some aspects of the plot. There are also a number of methods on the FacetGrid object for manipulating the figure at a higher level of abstraction. 

```python
with sb.axes_style("white"):
    g = sb.FacetGrid(tips,row="sex",col="smoker",margin_titles=True,height=2.5)
g.map(sb.scatterplot,"total_bill","tip",color="#334488")
g.set_axis_labels("Total bill($)","tip")
g.set(xticks=[10,30,50],yticks=[2,6,10])
g.fig.subplots_adjust(wspace=0.2,hspace=.2)
```

We can work directly with the underlining matplotlib Figure and Axes objects, which are stored as member attributes at fig and axes.  When making a figure without row or column faceting, we can also use the ax attributes to directly access the single axes.

```python
g=sb.FacetGrid(tips,col="smoker",margin_titles=True,height=4)
g.map(plt.scatter,"total_bill","tip",color="#338844",edgecolor="white",lw=1)
for ax in g.axes.flat:
    ax.axline((0,0),slope=.2,c=".2",ls="--",zorder=0)
g.set(xlim=(0,60),ylim=(0,14))
```

### ****Using Custom Functions

We are not limited to existing matplotlib and seaborn functions using FacetGrid. However, to work properly, any function we must follow a few rules 
1. It must plot onto the "currently active" matplotlib Axes.
2. It must accept the data that it plots in positional arguments. FacetGrid will pass a series of data for each of the named positional arguments passed to FacetGrid.map()
3. It must be able to accept color and label keyword arguments.In most cases, It's easiest to catch a generic dictionary of **kwargs and pass it along to the underlying plotting function.

```python
from scipy import stats
def quantile_plot(x,**kwargs):
    quantiles,xr=stats.probplot(x,fit=False)
    plt.scatter(xr,quantiles,**kwargs)
g=sb.FacetGrid(tips,col="sex",height=4)
g.map(quantile_plot,"total_bill")
```
