## Importing the libraries

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
sb.set_theme(style='darkgrid')
```

## Importing the dataset

```python
df = sb.load_dataset('tips')
df.head(5)
```

```python
sb.get_dataset_names()
```

## Scatterplot

```python
sb.relplot(x='total_bill',y='tip',data=df)
```

#####  We can also color the points based on class. In seaborn, this is referred to as using a “hue semantic”, because the color of the point gains meaning:

```python
sb.relplot(x="total_bill", y="tip", hue="smoker", data=df)
```

```python
sb.relplot(x="total_bill", y="tip", hue="sex", data=df)
```

##### To emphasize the difference between the classes, and to improve accessibility, We can use a different marker style for each class:

```python
sb.relplot(x="total_bill", y="tip", hue="smoker", style="smoker",
            data=df)
```

##### It’s also possible to represent four variables by changing the hue and style of each point independently. But this should be done carefully, because the eye is much less sensitive to shape than to color

```python
sb.relplot(x='total_bill',y='tip',hue = 'smoker',style = 'time',data=df)
```

##### In the examples above, the hue semantic was categorical, so the default qualitative palette was applied. If the hue semantic is numeric (specifically, if it can be cast to float), the default coloring switches to a sequential palette

```python
sb.relplot(x='total_bill',y='tip',hue='size',data=df)
```

##### In both cases, you can customize the color palette. There are many options for doing so. Here, we customize a sequential palette using the string interface to cubehelix_palette()

## Brief Introduction to cubehelix_palette()

#### seaborn.cubehelix_palette(n_colors=6, start=0, rot=0.4, gamma=1.0, hue=0.8, light=0.85, dark=0.15, reverse=False, as_cmap=False)
Make a sequential palette from the cubehelix system.This produces a colormap with linearly-decreasing (or increasing) brightness.
Parameters
    n_colors int
            Number of colors in the palette.

    start : float, 0 <= start <= 3
            The hue at the start of the helix.

    rot : float
            Rotations around the hue wheel over the range of the palette.
        
    gamma float 0 <= gamma
            Gamma factor to emphasize darker (gamma < 1) or lighter (gamma > 1) colors.

    hue float, 0 <= hue <= 1
            Saturation of the colors.

    dark float 0 <= dark <= 1
            Intensity of the darkest color in the palette.

    light float 0 <= light <= 1
            Intensity of the lightest color in the palette.

    reverse bool
            If True, the palette will go from dark to light.

    as_cmap bool
            If True, return a matplotlib.colors.Colormap.

    Returns
            list of RGB tuples or matplotlib.colors.Colormap

```python
sb.palplot(sb.cubehelix_palette())  #Default palette
```

```python
sb.palplot(sb.cubehelix_palette(rot=-0.4))
```

```python
sb.palplot(sb.cubehelix_palette(start=2.8,rot=0.1))
```

```python
sb.palplot(sb.cubehelix_palette(reverse=True))
```

```python
#Generating a colormap object
x=np.arange(25).reshape(5,5)
cmap = sb.cubehelix_palette(as_cmap=True)
ax = sb.heatmap(x, cmap=cmap)
```

```python
cmap = sb.cubehelix_palette(dark=0, light=1, as_cmap=True)
ax = sb.heatmap(x, cmap=cmap)
```

```python
sb.palplot(sb.color_palette("ch:2,r=.2,l=.6"))
```
