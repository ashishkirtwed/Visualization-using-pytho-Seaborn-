## Importing the libraries

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
sb.set_theme(style='darkgrid')
```

## Importing the dataset

```python
tips = sb.load_dataset('tips')
tips.head(5)
```

```python
sb.get_dataset_names()
```

## Scatterplot

```python
sb.relplot(x='total_bill',y='tip',data=tips)
```

#####  We can also color the points based on class. In seaborn, this is referred to as using a “hue semantic”, because the color of the point gains meaning:

```python
sb.relplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

```python
sb.relplot(x="total_bill", y="tip", hue="sex", data=tips)
```

##### To emphasize the difference between the classes, and to improve accessibility, We can use a different marker style for each class:

```python
sb.relplot(x="total_bill", y="tip", hue="smoker", style="smoker",
            data=tips)
```

##### It’s also possible to represent four variables by changing the hue and style of each point independently. But this should be done carefully, because the eye is much less sensitive to shape than to color

```python
sb.relplot(x='total_bill',y='tip',hue = 'smoker',style = 'time',data=tips)
```

##### In the examples above, the hue semantic was categorical, so the default qualitative palette was applied. If the hue semantic is numeric (specifically, if it can be cast to float), the default coloring switches to a sequential palette

```python
sb.relplot(x='total_bill',y='tip',hue='size',data=tips)
```

##### In both cases, you can customize the color palette. There are many options for doing so. Here, we customize a sequential palette using the string interface to cubehelix_palette()

## Brief Introduction to cubehelix_palette()

#### seaborn.cubehelix_palette(n_colors=6, start=0, rot=0.4, gamma=1.0, hue=0.8, light=0.85, dark=0.15, reverse=False, as_cmap=False)
Make a sequential palette from the cubehelix system.This produces a colormap with linearly-decreasing (or increasing) brightness.
Parameters
    n_colors int
            Number of colors in the palette.

    start : float, 0 <= start <= 3
            The hue at the start of the helix.

    rot : float
            Rotations around the hue wheel over the range of the palette.
        
    gamma float 0 <= gamma
            Gamma factor to emphasize darker (gamma < 1) or lighter (gamma > 1) colors.

    hue float, 0 <= hue <= 1
            Saturation of the colors.

    dark float 0 <= dark <= 1
            Intensity of the darkest color in the palette.

    light float 0 <= light <= 1
            Intensity of the lightest color in the palette.

    reverse bool
            If True, the palette will go from dark to light.

    as_cmap bool
            If True, return a matplotlib.colors.Colormap.

    Returns
            list of RGB tuples or matplotlib.colors.Colormap

```python
sb.palplot(sb.cubehelix_palette())  #Default palette
```

```python
sb.palplot(sb.cubehelix_palette(rot=-0.4))
```

```python
sb.palplot(sb.cubehelix_palette(start=2.8,rot=0.1))
```

```python
sb.palplot(sb.cubehelix_palette(reverse=True))
```

```python
#Generating a colormap object
x=np.arange(25).reshape(5,5)
cmap = sb.cubehelix_palette(as_cmap=True)
ax = sb.heatmap(x, cmap=cmap)
```

```python
cmap = sb.cubehelix_palette(dark=0, light=1, as_cmap=True)
ax = sb.heatmap(x, cmap=cmap)
```

## Emphasizing continuity with line plots

###### lineplot() assumes that you are most often trying to draw y as a function of x, the default behaviour is to sort the data by the x values before plotting.

```python
df = pd.DataFrame(dict(time = np.arange(500),
                      value = np.random.randn(500).cumsum()))
g = sb.relplot(x="time",y="value",kind="line",data=df)
g.fig.autofmt_xdate()
```

```python
df = pd.DataFrame(np.random.randn(500,2).cumsum(axis=0),columns=["x","y"])
sb.relplot(x="x",y="y",sort=False,kind="line",data=df)
```

#### Aggregating and representing uncertainty

#### More complex datasets will have multiple measurements for the same value of the x variable. The default behavior in seaborn is to aggregate the multiple measurements at each x value by plotting the mean and the 95% confidence interval around the mean

```python
fmri = sb.load_dataset("fmri")
fmri.head(5)
```

```python
sb.relplot(x="timepoint",y="signal",kind="line",data=fmri)
```

```python
sb.relplot(x="timepoint",y="signal",kind="line",ci="sd",data=fmri)
```

#### Plotting subsets of data with semantic mappings 

#### Adding a hue semantic with two levels splits the plot into two lines and error bands, coloring each to indicate which subset of the data they correspond to.

```python
sb.relplot(x="timepoint",y="signal",hue = "event",kind="line",data=fmri)
```

```python
sb.relplot(x="timepoint",y="signal",hue="region",style="event",kind="line",data=fmri)
```

#### Plotting with date data

###### Line plots are often used to visualize data associated with real dates and times. These functions pass the data down in their original format to the underlying matplotlib functions, and so they can take advantage of matplotlib’s ability to format dates in tick labels.

```python
df = pd.DataFrame(dict(time=pd.date_range("2017-1-1", periods=500),
                       value=np.random.randn(500).cumsum()))
g = sb.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()
```

### Introduction to the FacetGrid

##### Multi-plot grid for plotting conditional relationships

### class seaborn.FacetGrid(**kwargs)

```python
sb.FacetGrid(tips)
```

#### Assign column and/or row variables to add more subplots to the figure:

```python
sb.FacetGrid(tips, col="time", row="sex")
```

#### To draw a plot on every facet, pass a function and the name of one or more columns in the dataframe to FacetGrid.map()

```python
g = sb.FacetGrid(tips, col="time",  row="sex")
g.map(sb.scatterplot, "total_bill", "tip")
```

#### The variable specification in FacetGrid.map() requires a positional argument mapping, but if the function has a data parameter and accepts named variable assignments, you can also use FacetGrid.map_dataframe()

```python
g = sb.FacetGrid(tips, col="time",  row="sex")
g.map_dataframe(sb.histplot, x="total_bill")
```

#### One difference between the two methods is that FacetGrid.map_dataframe() does not add axis labels. There is a dedicated method to do that

```python
g = sb.FacetGrid(tips, col="time",  row="sex")
g.map_dataframe(sb.histplot, x="total_bill")
g.set_axis_labels("Total bill", "Count")
```

#### setting binwidth to facetgrid

```python
g = sb.FacetGrid(tips, col="time", row="sex")
g.map_dataframe(sb.histplot, x="total_bill", binwidth=2)
g.set_axis_labels("Total bill", "Count")
```

#### The FacetGrid constructor accepts a hue parameter. Setting this will condition the data on another variable and make multiple plots in different colors. Where possible, label information is tracked so that a single legend can be drawn

```python
g = sb.FacetGrid(tips, col="time", hue="sex")
g.map_dataframe(sb.scatterplot, x="total_bill", y="tip")
g.set_axis_labels("Total bill", "Tip")
g.add_legend()
```

#### The size and shape of the plot is specified at the level of each subplot using the height and aspect parameters: Change the height and aspect ratio of each facet

```python
g = sb.FacetGrid(tips, col="day", height=3.5, aspect=.65)
g.map(sb.histplot, "total_bill")
```

#### You can pass custom functions to plot with, or to annotate each facet. Your custom function must use the matplotlib state-machine interface to plot on the “current” axes, and it should catch additional keyword arguments:

```python
import matplotlib.pyplot as plt
def annotate(data, **kws):
    n = len(data)
    ax = plt.gca() # gca() get current axis
    ax.text(.1, .6, f"N = {n}", transform=ax.transAxes)

g = sb.FacetGrid(tips, col="time")
g.map_dataframe(sb.scatterplot, x="total_bill", y="tip")
g.set_axis_labels("Total bill", "Tip")
g.map_dataframe(annotate)
```
